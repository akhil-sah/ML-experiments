## Aim : Linear Regression using normal equation.

### Import dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

### Preprocess input data 

```python
#read csv file
df = pd.read_csv("data.csv")

#extract x co-ordinates of all the points (independent variable)
X = df.iloc[:,0]
x_ = df.iloc[:,[0]]
x_['x0'] = [1]*len(X)
temp = x_.pop('x0')
x_.insert(0, 'x0', temp)

#extract x co-ordinates of all the points (dependent variable)
Y = df.iloc[:,1]

#to obtain a summary of data set
df.info()

#plotting the data values
plt.style.use("seaborn")
plt.figure()
plt.suptitle("Marks")
plt.xlabel("SAT score")
plt.ylabel("GPA")
plt.scatter(X, Y)
plt.show()
```

### Normal equation to find weight parameters

```python
#converting pandas dataframe into numpy arrays
x = np.c_[x_]
y = np.c_[Y]
print(x, y)

#normal equation
theta = np.linalg.inv(x.T.dot(x)).dot(x.T).dot(y)
print(theta)
```

```python
#Making predictions
y_pred = x.dot(theta)
print("Coefficient of determination: ", r2_score(Y, y_pred))
plt.scatter(X, Y, label = "Actual")
plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color='red', label = "Predicted")  # regression line
plt.legend()
plt.show()
```

The best fitting curve obtained for given data set is as obtained in above plot. The scatter indicates that there is a weak positive correlation between GPA and SAT score. The best fit obtained also indicates the same.
The regression line obtained is almost same as the one obtained from using the gradient descent algorithm.

### Linear Regression for multivariate function

```python
#Preprocess the dataset
df = pd.read_csv("d_set.csv")
X = df.iloc[:, [0, 1, 2]]
Y = df.iloc[:, [3]]

#adding 
X['x0'] = [1]*99
temp = X.pop('x0')
X.insert(0, 'x0', temp)
print(X)
print(Y)
```

```python
#convert pandas dataframe into numpy array
x = np.c_[X]
y = np.c_[Y]
print(x, y)
```

```python
#Normal equation
theta = np.linalg.inv(x.T.dot(X)).dot(x.T).dot(y)
print(theta)
y_pred = x.dot(theta)
#R^2 score (measure of accuracy of model)
print("Coefficient of Determination:", r2_score(Y, y_pred))
```
