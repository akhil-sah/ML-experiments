## Aim : Write a Python program for Linear Regression with Gradient Descent algorithm. 

### Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Preprocess input data

```python
#read csv file
df = pd.read_csv("data.csv")

#extract x co-ordinates of all the points (independent variable)
X = df.iloc[:,0]

#extract x co-ordinates of all the points (dependent variable)
Y = df.iloc[:,1]

#to obtain a summary of data set
df.info()

#plotting the data values
plt.style.use("seaborn")
plt.figure()
plt.suptitle("Marks")
plt.xlabel("SAT score")
plt.ylabel("GPA")
plt.scatter(X, Y)
plt.show()
```

### Linear Regression Model

```python
#intialising m and c with 0
m = 0. 
c = 0.

alpha = 0.0000001  # learning rate
epochs = 1000      # number of iterations to perform gradient descent
n = float(len(X))  # number of elements in data set 

for i in range(epochs):
    #predict value of Y on basis of current value of m and c
    Y_pred = m * X + c    
    
    #differentiate cost function w.r.t. m
    dm = (-2/n) * sum(X * (Y - Y_pred))
    
    #differentiate cost function w.r.t. c
    dc = (-2/n) * sum(Y - Y_pred)
    
    #update m and c
    m = m - alpha * dm
    c = c - alpha * dc

print(m, c)
```

After runing the gradient descent algorithm we obtained the optimum values of m and c as 0.0018042675296552377 and 1.148553038683468e-06 respectively.

```python
#Making predictions
Y_pred = m * X + c
plt.scatter(X, Y, label = "Actual")
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red', label = "Predicted")  # regression line
plt.legend()
plt.show()
```

The best fitting curve obtained for given data set is as obtained in above plot. The scatter indicates that there is a weak positive correlation between GPA and SAT score. The regression line obtained also indicates the same.
